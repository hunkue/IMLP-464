```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
# 在jupyter run 的時候可以直接秀出來，現在的 jypyter 應該不用輸入了
```

```python
days = range(1, 7)
celsius_max = [27.6, 26.1, 22.5, 30.4, 29.5, 31.5]
celsius_min = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5]
plt.plot(days, celsius_max, "o-r",
         days, celsius_min, "^--g")
xmin, xmax, ymin, ymax = 0, 7, 0, 40
plt.axis([xmin, xmax, ymin, ymax])
plt.xlabel("day")
plt.ylabel("temperature (unit: Celsius)")
plt.title("daliy average temperature")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "m-o", label = "sin(x)")
plt.plot(x, cosinus, "c--s", label = "cos(x)")
plt.legend(loc=1)
plt.show()
```

```python
# 機器學習生成亂數可能會需要參考
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```

```python
x = np.linspace(0,10,50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.subplot(2, 1, 1)
plt.plot(x, sinus, "r-o")
plt.subplot(2, 1, 2)
plt.plot(x, cosinus, "g--")
plt.show()
```

```python
x = np.linspace(0,10,50)
sinus = np.sin(x)
sinhs = np.sinh(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, sinus, "r-o")
ax2 = ax.twinx()
ax2.plot(x, sinhs, "g--")
plt.show()
```
